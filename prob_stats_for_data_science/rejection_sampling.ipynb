```python
import numpy as np
import matplotlib.pyplot as plt

u1 = np.random.rand(10000)
u2 = np.random.rand(10000)

x = np.zeros(1000)
y = np.zeros(1000)
i = 0
j = 0
while i < 1000:
    if (u1[j] < 0.5) and (u2[j] <= 2*u1[j]):
        x[i] = u1[j]
        y[i] = u2[j]
        i+=1
        j+=1
    elif (u1[j] > 0.5) and (u2[j] <= -2*u1[j] + 2):
        x[i] = u1[j]
        y[i] = u2[j]
        i+=1
        j+=1
    else:
        j+=1

fig = plt.figure()
plt.scatter(x,y)
plt.show()
```
