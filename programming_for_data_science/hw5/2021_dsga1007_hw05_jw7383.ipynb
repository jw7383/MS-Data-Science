**Problem 1.9 (Points 20)**

Normalize each of the three columns `math score`, `reading score`, `writing score` by subtracting its mean value and dividing by its standard deviation, and form three new columns: `math_score_norm`, `reading_score_norm`, `writing_score_norm`, respectively.

Perform the same task the one above (1.8) on these newly formed columns.  


```python
df['math_score_norm'] = (df['math score'] - df['math score'].mean())/df['math score'].std()
df['reading_score_norm'] = (df['reading score'] - df['reading score'].mean())/df['reading score'].std()
df['writing_score_norm'] = (df['writing score'] - df['writing score'].mean())/df['writing score'].std()
df.plot.scatter(x = 'math_score_norm', y = 'reading_score_norm', alpha = 0.5)
df.plot.scatter(x = 'math_score_norm', y = 'writing_score_norm', alpha = 0.5)
df.plot.scatter(x = 'reading_score_norm', y = 'writing_score_norm', alpha = 0.5)
```
