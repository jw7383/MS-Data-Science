**1.10 (15 points)**

Use the ratings `rating_final.csv`. 'Group by' the data by `placeID` and obtain `mean` and `std` of each of the ratings (`rating`, `food_rating`, and `service_rating`) for each restaurant (i.e. for each `placeID`).

Print the head of this frame so to show the results. 

```python
rating_final.groupby('placeID')[['rating','food_rating','service_rating']].agg(['mean', 'std']).head()
```

**1.11 (10 points)**


Rank the restaurants IDs (i.e. 'placeID') by their `mean ratings` increasingly.

Print the restaurants IDs (i.e. 'placeID') and `mean ratings` with the lowest `mean ratings` (hint: lowest appears at the top).

Print the restaurants IDs (i.e. 'placeID') and `mean ratings` with the highest `mean ratings` (hint: highest appears at the bottom).



```python
mean_ratings = rating_final.groupby('placeID')['rating'].mean()
mean_ratings_sorted = mean_ratings.sort_values()
print('Lowest mean rating by placeID:')
print(mean_ratings_sorted.head(1))
print('Highest mean rating by placeID:')
print(mean_ratings_sorted.tail(1))
```

**1.12 (10 points)**

Make and print a table with columns (`place`, `mean`, `name`), 
which are the restaurant's ID, mean ratings, and name, 
of the 10 restaurants with the lowest `mean ratings`.


```python
dflow = pd.merge(mean_ratings, geoplaces2, on = 'placeID')
dflow = dflow.sort_values(by = 'rating')
dflow = dflow[['placeID', 'name', 'rating']].head(10)
dflow.columns = ['place', 'name', 'mean']
dflow
```

**1.13 (10 points)**

Make and print a table with columns (`place`, `mean`, `name`), 
which are the restaurant's ID, mean ratings, and name, 
of the 10 restaurants with the highest `mean ratings`.


```python
dfhigh = pd.merge(mean_ratings, geoplaces2, on = 'placeID')
dfhigh = dfhigh.sort_values(by = 'rating')
dfhigh = dfhigh[['placeID', 'name', 'rating']].tail(10)
dfhigh.columns = ['place', 'name', 'mean']
dfhigh
```

**1.14 (10 points)**


Print `name`, `city`, `state` for the two restaurants with the lowest and highest rank from the "1.11" problem sorted data frame (which ranks the restaurants IDs (i.e. 'placeID') by their `mean ratings` increasingly).




```python
dflowhigh = pd.merge(mean_ratings, geoplaces2, on = 'placeID')
dflowhigh = dflowhigh.sort_values(by = 'rating')
ncslow = dflowhigh[['name', 'city', 'state']].head(1)
ncshigh = dflowhigh[['name', 'city', 'state']].tail(1)
print('Name, city, and state for the restaurant with the lowest mean rating:')
print(ncslow)
print('-'*80)
print('Name, city, and state for the restaurant with the highest mean rating:')
print(ncshigh)
```
